```python
import pandas as pd, numpy as np, ast, matplotlib.pyplot as plt

# Load and parse the CSV
PATH = "/workspace/text2program-for-ehr/saved/ensemble/decode_outputs/ensemble_test_result_beam_5_num-5.csv"
df = pd.read_csv(PATH, converters={"data_uncertainty": ast.literal_eval})
df["recover_ex_flag"] = df["recover_ex_flag"].astype(bool)

# Compute mean data uncertainty per prediction
df["row_uncertainty"] = df["data_uncertainty"].apply(np.mean)

# Compute mean uncertainty per question
q_unc = df.groupby("question")["row_uncertainty"].mean()
df = df.merge(q_unc.rename("q_unc"), left_on="question", right_index=True)
questions_sorted = q_unc.sort_values(ascending=False).index.tolist()
num_q = len(questions_sorted)

# Accuracy function with percent-based top-N
def accuracy(pct_uncertain):
    top_N = int(pct_uncertain * num_q)
    uncertain_set = set(questions_sorted[:top_N])
    correct = 0
    by_q = df.groupby("question")
    for q, g in by_q:
        if q in uncertain_set:
            correct += g["recover_ex_flag"].any()
        else:
            correct += g.loc[g["idx"].idxmin(), "recover_ex_flag"]
    return correct / num_q

# Sweep percentage from 0% to 100%
percent_vals = np.linspace(0, 1, 101)
results = [{"Percent": int(p*100), "Accuracy": accuracy(p)} for p in percent_vals]
res_df = pd.DataFrame(results)

# Plot
plt.plot(res_df["Percent"], res_df["Accuracy"], marker="o")
plt.xlabel("Top % most-uncertain questions given 5 predictions")
plt.ylabel("Overall accuracy (using recover_ex_flag)")
plt.title("Accuracy vs. adaptive beam allocation (percent-based)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
